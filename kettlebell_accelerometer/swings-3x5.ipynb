```
import pandas as pd

url = 'https://raw.githubusercontent.com/emh/emh.github.io/master/kettlebell_accelerometer/swings-3x5.csv'
data = pd.read_csv(url)
data = data[["Date", "G"]]
data.head()
```

```
data.rename(columns = {'G': 'a', 'Date': 't'}, inplace = True)
data['a'] = data['a'].apply(lambda g: (g - 1) * -9.8) # remove gravity and convert to m/s^2
t0 = data['t'][0]
data['t'] = data['t'].apply(lambda t: (pd.Timestamp(t) - pd.Timestamp(t0)).total_seconds())
data.head()
```

```
import altair as alt

alt.Chart(data).mark_line().encode(
    x='t',
    y='a'
)
```
